# Checkpoint 5: Sector Analysis 2

```python
D1 = master_frame[(master_frame['country_code'] == 'USA') & 
             (master_frame['raised_amount_usd'] >= 5) & 
             (master_frame['raised_amount_usd'] <= 15)]
```

```python
D1_value = D1[['raised_amount_usd','main_sector']].groupby('main_sector').agg(['sum', 'count']).rename(
    columns={'sum':'Total_amount','count' : 'Total_count'})
D1 = D1.merge(D1_value, how='left', on ='main_sector')
D1.head()
```

```python
D1.info()
```

```python
# D2 contains the values of GBR FUNDS having investment between 5-15 million
#The total number (or count) of investments for each main sector in a separate column
#The total amount invested in each main sector in a separate column
D2 = master_frame[(master_frame['country_code'] == 'GBR') & 
             (master_frame['raised_amount_usd'] >= 5) & 
             (master_frame['raised_amount_usd'] <= 15)]
D2_value = D2[['raised_amount_usd','main_sector']].groupby('main_sector').agg(['sum', 'count']).rename(
    columns={'sum':'Total_amount','count' : 'Total_count'})
D2 = D2.merge(D2_value, how='left', on ='main_sector')
D2.head()
```

```python
# D2 contains the values of GBR FUNDS having investment between 5-15 million
#The total number (or count) of investments for each main sector in a separate column
#The total amount invested in each main sector in a separate column
D3 = master_frame[(master_frame['country_code'] == 'IND') & 
             (master_frame['raised_amount_usd'] >= 5) & 
             (master_frame['raised_amount_usd'] <= 15)]
D3_value = D3[['raised_amount_usd','main_sector']].groupby('main_sector').agg(['sum', 'count']).rename(
    columns={'sum':'Total_amount','count' : 'Total_count'})
D3 = D3.merge(D3_value, how='left', on ='main_sector')
D3.head()
```

```python
#The total number (or count) of investments for each country
print(D1.raised_amount_usd.count())
print(D2.raised_amount_usd.count())
print(D3.raised_amount_usd.count())
```

```python
#The total SUM investments for each country
print(round(D1.raised_amount_usd.sum(), 2))
print(round(D2.raised_amount_usd.sum(), 2))
print(round(D3.raised_amount_usd.sum(), 2))
```

```python
#For the top and second top sector GBR , which company received the highest investment?
company2 = D2[D2['main_sector']=='Others']
company2 = company2.pivot_table(values ='raised_amount_usd', index = 'company_permalink', aggfunc = 'sum')
company2 = company2.sort_values(by = 'raised_amount_usd', ascending = False).head()
print(company2.head(2))
```

```python
#For the top and second top sector IND , which company received the highest investment?
company3 = D3[D3['main_sector']=='Others']
company3 = company3.pivot_table(values = 'raised_amount_usd', index = 'company_permalink', aggfunc = 'sum')
company3 = company3.sort_values(by = 'raised_amount_usd', ascending = False).head()
print(company3.head(2))
```
